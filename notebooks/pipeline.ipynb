```python
from d2c.data_generation.builder import TSBuilder
from d2c.descriptors import D2C, DataLoader

maxlags=3
n_variables=3

tsbuilder = TSBuilder(observations_per_time_series=50, 
                    maxlags=maxlags, 
                    n_variables=n_variables, 
                    time_series_per_process=1, 
                    processes_to_use=list(range(1, 21)), 
                    noise_std=0.1, 
                    max_neighborhood_size = 4, 
                    seed=1, 
                    verbose=True)

tsbuilder.build()

dataloader = DataLoader(n_variables = n_variables,
                        maxlags = maxlags)
dataloader.from_tsbuilder(tsbuilder)
```

```python
d2c = D2C(observations=dataloader.get_observations(), 
        dags=dataloader.get_dags(), 
        couples_to_consider_per_dag=60, 
        MB_size=5, 
        n_variables=n_variables, 
        maxlags=maxlags,
        seed=17,
        n_jobs=25)

d2c.initialize()

dd1 = d2c.get_descriptors_df()
```

```python
from d2c.data_generation.builder import TSBuilder
from d2c.descriptors import D2C, DataLoader

maxlags=3

for n_variables in [5]:
    tsbuilder = TSBuilder(observations_per_time_series=250, 
                        maxlags=maxlags, 
                        n_variables=n_variables, 
                        time_series_per_process=100, 
                        processes_to_use=list(range(1, 21)), 
                        noise_std=0.1, 
                        max_neighborhood_size = 4, 
                        seed=n_variables, 
                        verbose=True)

    tsbuilder.build()

    dataloader = DataLoader(n_variables = n_variables,
                            maxlags = maxlags)
    dataloader.from_tsbuilder(tsbuilder)

    d2c = D2C(observations=dataloader.get_observations(), 
            dags=dataloader.get_dags(), 
            couples_to_consider_per_dag=60, 
            MB_size=5, 
            n_variables=n_variables, 
            maxlags=maxlags,
            n_jobs=25)

    d2c.initialize()

    descriptors = d2c.get_descriptors_df()
    descriptors.to_pickle(f'../data/jacopo.pkl')
```
