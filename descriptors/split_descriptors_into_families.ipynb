```python
import pickle as pkl

```

```python
folder = '../2024.01.20/'
with open(folder+'descriptors_Ridge-basic-var_var-var_mb-var_mb_given_var-mb_mb_given_var-structural.pkl', 'rb') as f:
    descriptors = pkl.load(f)
```

```python
descriptors.shape
```

```python
pq= [0.05,0.1,0.25,0.5,0.75,0.9,0.95]
family = {}
family['index'] = ['graph_id','edge_source','edge_dest']
family['var_var'] = ["coefficient_ca", 
                    "coefficient_ef", 
                    "ca_ef_given_common_causes", 
                    "ef_ca_given_MBef", 
                    "ca_ef_given_MBca"] + \
                    [f"ef_ca_given_MBca_plus_MBef_{i}" for i in range(len(pq))] + \
                    [f"ca_ef_given_MBef_plus_MBca_{i}" for i in range(len(pq))] +  \
                    ["ca_ef", "ef_ca"]

family['var_mb'] = [f"MBef_ca_{i}" for i in range(len(pq))] + [f"MBca_ef_{i}" for i in range(len(pq))]

family['var_mb_given_var'] = [f"ca_MBef_given_ef_{i}" for i in range(len(pq))] + [f"ef_MBca_given_ca_{i}" for i in range(len(pq))]

family['mb_mb_given_var'] = [f"MBca_MBef_given_ca_{i}" for i in range(len(pq))] + [f"MBca_MBef_given_ef_{i}" for i in range(len(pq))]

family['structural'] = [f"Structural_1_{i}" for i in range(len(pq))] + [f"Structural_2_{i}" for i in range(len(pq))]

family['basic'] =  ["n_observations", 
                "n_features", 
                "n_features/n_observations", 
                "kurtosis_ca", 
                "kurtosis_ef", 
                "skewness_ca", 
                "skewness_ef",
                "HOC_1_2",
                "HOC_2_1",
                "HOC_1_3",
                "HOC_3_1"]
```

```python
assert( descriptors.shape[1] == len(family['index']) + len(family['var_var']) + len(family['var_mb']) + len(family['var_mb_given_var']) + len(family['mb_mb_given_var']) + len(family['structural']) + len(family['basic']) + 1) #+1 for label
```

```python
#make one big list from all the lists
all_descriptors = []
for key in family.keys():
    all_descriptors += family[key]

# replace the column names
descriptors.columns = all_descriptors + ['is_causal']
```

```python
# create folder folder +'families/' 
import os
if not os.path.exists(folder + 'families/'):
    os.makedirs(folder + 'families/')
    
```

```python
# make family couples 
family_couples = []
for i in range(1,len(family.keys())):
    for j in range(i+1, len(family.keys())):
        family_couples.append([list(family.keys())[i],list(family.keys())[j]])

# make family triplets
family_triplets = []
for i in range(1,len(family.keys())):
    for j in range(i+1, len(family.keys())):
        for k in range(j+1, len(family.keys())):
            family_triplets.append([list(family.keys())[i],list(family.keys())[j],list(family.keys())[k]])

# make family quadruplets
family_quadruplets = []
for i in range(1,len(family.keys())):
    for j in range(i+1, len(family.keys())):
        for k in range(j+1, len(family.keys())):
            for l in range(k+1, len(family.keys())):
                family_quadruplets.append([list(family.keys())[i],list(family.keys())[j],list(family.keys())[k],list(family.keys())[l]])

# make family quintuplets
family_quintuplets = []
for i in range(1,len(family.keys())):
    for j in range(i+1, len(family.keys())):
        for k in range(j+1, len(family.keys())):
            for l in range(k+1, len(family.keys())):
                for m in range(l+1, len(family.keys())):
                    family_quintuplets.append([list(family.keys())[i],list(family.keys())[j],list(family.keys())[k],list(family.keys())[l],list(family.keys())[m]])

# make family sextuplets
family_sextuplets = []
for i in range(1,len(family.keys())):
    for j in range(i+1, len(family.keys())):
        for k in range(j+1, len(family.keys())):
            for l in range(k+1, len(family.keys())):
                for m in range(l+1, len(family.keys())):
                    for n in range(m+1, len(family.keys())):
                        family_sextuplets.append([list(family.keys())[i],list(family.keys())[j],list(family.keys())[k],list(family.keys())[l],list(family.keys())[m],list(family.keys())[n]])
```

```python
# save the descriptors for each family
for family_index, key in enumerate(family.keys()):
    if key == 'index':
        continue
    descriptors[family['index'] + family[key] + ['is_causal']].to_pickle(folder + 'families/'+ 'descriptors_Ridge_f' + str(family_index) + '.pkl')

# save the descriptors for each family couple/triplet/quadruplet/quintuplet/sextuplet
for family_couple in family_couples+family_triplets+family_quadruplets+family_quintuplets+family_sextuplets:
    list_descriptors = [family[current_element] for current_element in family_couple]
    flattened_list_descriptors = [item for sublist in list_descriptors for item in sublist]
    list_indexes = [str(list(family.keys()).index(current_element)) for current_element in family_couple]
    list_indexes_prepared = str(list_indexes).strip('[]').replace(', ','-').replace('\'','')
    descriptors[family['index'] + flattened_list_descriptors  + ['is_causal']].to_pickle(folder + 'families/'+ 'descriptors_Ridge_f' + list_indexes_prepared  + '.pkl')

```
